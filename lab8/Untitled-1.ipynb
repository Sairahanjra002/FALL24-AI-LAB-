```python
import math

# Improvement added: Alpha-Beta pruning to skip unnecessary branches
def minimax(depth, index, isMax, values, maxDepth, alpha=-math.inf, beta=math.inf):

    if depth == maxDepth:
        return values[index]

    if isMax:
        best = -math.inf
        
        v1 = minimax(depth + 1, index * 2, False, values, maxDepth, alpha, beta)
        best = max(best, v1)
        alpha = max(alpha, best)
        if alpha >= beta:
            return best

        v2 = minimax(depth + 1, index * 2 + 1, False, values, maxDepth, alpha, beta)
        best = max(best, v2)
        alpha = max(alpha, best)

        return best

    else:
        best = math.inf
        
        v1 = minimax(depth + 1, index * 2, True, values, maxDepth, alpha, beta)
        best = min(best, v1)
        beta = min(beta, best)
        if alpha >= beta:
            return best

        v2 = minimax(depth + 1, index * 2 + 1, True, values, maxDepth, alpha, beta)
        best = min(best, v2)
        beta = min(beta, best)

        return best


values = [3, 5, 2, 9, 3, 5, 2, 9]
maxDepth = int(math.log(len(values), 2))

print("The optimal value is:", minimax(0, 0, True, values, maxDepth))
```
